```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/shamnad-sherief/datasets/main/50_Startups.csv')
```

```python
x = dataset.iloc[:,:4].values
```

```python
y = dataset.iloc[:,4:5].values
```

```python
# Converting type of columns to category
dataset['State']=dataset['State'].astype('category')
```

```python
#Assigning numerical values and storing it in another columns
dataset['Stat_new']=dataset['State'].cat.codes
```

```python
dataset
```

```python
#Create an instance of One-hot-encoder
enc=OneHotEncoder()
```

```python
#Passing encoded columns
enc_data=pd.DataFrame(enc.fit_transform(dataset[['Stat_new']]).toarray())
```

```python
enc_data
```

```python
#Merge with main
New_df=dataset.join(enc_data)
```

```python
New_df
```

```python
one_hot_encoded_data = pd.get_dummies(dataset, columns = ['State'])
print(one_hot_encoded_data)
```

```python
x_variable = New_df.drop(New_df.columns[[3]], axis = 1)
```

```python
x_variable = New_df.drop(["State", "Stat_new"], axis = 1)
```

```python
x_variable
```

```python
from pandas.core.common import random_state
#spiltting the dataset into training and test set.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_variable, y, test_size= 0.3)
```

```python
# Fitting the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
```

```python
score
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
import pickle
# create an iterator object with write permission - model.pkl
with open('model_pkl', 'wb') as files:
    pickle.dump(regressor, files)
```
